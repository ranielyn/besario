Exercise 1: Regular Expressions
Load the Shakespeareâ€™s Hamlet from the NLTK (shakespeare-hamlet.txt)

Words containing 2 vowels characters

```python
import re
import nltk
import math
import string

nltk.download('gutenberg')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
hamlet_raw = nltk.corpus.gutenberg.raw('shakespeare-hamlet.txt') #raw hamlet

re.findall(r'(?<=\W)[A-Za-z][aeiou]{2}[A-Za-z](?=\W)',hamlet_raw)
```

Words that contain a hyphen

```python
import re
re.findall(r'\w+-\w+[-\w+]*',hamlet_raw)
```

Words in their present progressive form (ends in -ing)

```python
re.findall(r'(?<=\W)[A-Za-z-]+ing(?=\W)',hamlet_raw)
```

Words in their superlative form (ends in -est)

```python
re.findall(r'(?<=\W)[A-Za-z-]+est(?=\W)',hamlet_raw)
```

Words at the end of the sentence (next to sentence ending punctuations)

```python
re.findall(r'(?<=[?.!]\W)(?<=\W)[A-Z][a-z-]+(?=\W)',hamlet_raw)
```

Tokenize the data using Penn Tree Bank Tokenization

```python
ptb_tokenizer = nltk.tokenize.treebank.TreebankWordTokenizer()
PTB_TOKENS = ptb_tokenizer.tokenize(hamlet_raw)
PTB_TOKENS[:500]
```

Apply case folding on the tokens

```python
CF_PTB_TOKENS = [w.lower() for w in PTB_TOKENS]
CF_PTB_TOKENS[:500]
```

Lemmatize the case folded tokens using Word Net Lemmatizer

```python
import nltk
nltk.download('wordnet')
lemmatizer = nltk.WordNetLemmatizer()
WNL_PTB_TOKENS = [lemmatizer.lemmatize(w) for w in PTB_TOKENS]
WNL_PTB_TOKENS[:500]
```
